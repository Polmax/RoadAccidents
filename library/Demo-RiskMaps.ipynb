## RiskMap demo notebook
### This notebook contains examples on how to use different features from the RiskMap library,the examples used to showcase this features in two GeoJson files containing the specifications of two urban areas in the portuguese districts of Braga and Vila Real

```python
from RiskMaps import RiskMapFromRiskLocations,RiskMapFromLocation,RiskLocationsFromLocation,load_models_and_settings
import cv2
import matplotlib.pyplot as plt
import json
```

```python
def show_image(path,title):
  f,ax = plt.subplots(1,1,figsize=(24,8))
  im = cv2.imread(path)
  ax.imshow(im)
  ax.axis('off')
  f.suptitle(title)
  plt.show()
```

#### load already trained models and other setting into memory

```python
load_models_and_settings(setting_file="settings.json")
```

### Use of the RiskMap library with the Braga GeoFile

#### Aerial image from a region in the district of Braga

```python
show_image("test_files/braga_satellite.png","sattelite braga")
```

#### Map from the same region

```python
show_image("test_files/braga_mapbox.png","map braga")
```

#### Write the risk locations to a geojson file

```python
GeoFile = "test_files/braga_example.geojson"
size = (1024,1024)

RiskLocationsFromLocation(GeoFile,out_loc="risk_braga.geojson",size=size,accuracy="detailed")
```

```python
with open("risk_braga.geojson") as f:
    risk = json.load(f)
    
print(json.dumps(risk,indent = 4))
```

#### Create the risk map using the risk locations file created before

```python
RiskMapFromRiskLocations(GeoFile,risks_loc="risk_braga.geojson",out_loc="risk_braga.tif",size=size,accuracy="detailed")
```

```python
show_image("risk_braga.tif","RiskMap braga")
```

#### Create the risk map for the same location

```python
RiskMapFromLocation(GeoFile,out_loc="risk_braga.tif",size=size,accuracy="detailed")
```

```python
show_image("risk_braga.tif","RiskMap braga")
```
