```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import random
```

```python
SOURCE_PATH = "images/portugal"
DEST_PATH = "resized_images/portugal"
try:
    os.makedirs("resized_images/portugal")
except:
    print("file already exists")
    
models = ["MODEL1","MODEL2"]

categories = [['ACCIDENTS','SAFE'],['LEVEL1','LEVEL2','LEVEL3']]
```

```python
for m in range(len(models)):
    model = models[m]
    try:
        os.makedirs(f"{DEST_PATH}/{model}")
    except:
        print("file already exists")
        
    for c in range(len(categories[m])):
        categorie = categories[m][c]
        try:
            os.makedirs(f"{DEST_PATH}/{model}/{categorie}")
        except:
            print("file already exists")
        imgs = os.listdir(f"{SOURCE_PATH}/{model}/{categorie}")
        sample_normal = random.sample(imgs,6)
        print(sample_normal)
        f,ax = plt.subplots(2,3,figsize=(15,9))

        for i in range(0,6):
            im = cv2.imread(f'{SOURCE_PATH}/{model}/{categorie}/{sample_normal[i]}')
            im = im[:475]
            ##im = cv2.resize(im,(128,128))
            ax[i//3,i%3].imshow(im)
            ax[i//3,i%3].axis('off')
        f.suptitle(f'{categorie.upper()}')
        plt.show()
        
        for i in imgs:
            im = cv2.imread(f'{SOURCE_PATH}/{model}/{categorie}/{i}')
            im = im[:475]
            new_name = i.replace("png","jpg")
            im = cv2.resize(im,(128,128))
            ##cv2.imwrite('image.jpg', image,
             
            cv2.imwrite(f"{DEST_PATH}/{model}/{categorie}/{new_name}",im,[cv2.IMWRITE_JPEG_QUALITY, 30])
            
     
```
