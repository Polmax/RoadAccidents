```python
geo_json_geometry = {
  "type": "Polygon",
  "coordinates": [
    [
      [
        -122.52227783203125,
        40.660847697284815
      ],
      [
        -122.52227783203125,
        40.987154933797335
      ],
      [
        -122.01690673828124,
        40.987154933797335
      ],
      [
        -122.01690673828124,
        40.660847697284815
      ],
      [
        -122.52227783203125,
        40.660847697284815
      ]
    ]
  ]
}
```

```python
geometry_filter = {
  "type": "GeometryFilter",
  "field_name": "geometry",
  "config": geo_json_geometry
}

date_range_filter = {
  "type": "DateRangeFilter",
  "field_name": "acquired",
  "config": {
    "gte": "2016-07-01T00:00:00.000Z",
    "lte": "2016-08-01T00:00:00.000Z"
  }
}

cloud_cover_filter = {
  "type": "RangeFilter",
  "field_name": "cloud_cover",
  "config": {
    "lte": 0.1
  }
}
redding_reservoir = {
  "type": "AndFilter",
  "config": [geometry_filter,date_range_filter,cloud_cover_filter]
}
```

```python
import os
import requests
from requests.auth import HTTPBasicAuth
import json

API_KEY = "0c2fc5b9aa344af5be9991ac04188962"
# our demo filter that filters by geometry, date and cloud cover

search_endpoint_request = {
  "item_types": ["REOrthoTile"],
  "filter": redding_reservoir
}

# fire off the POST request
result = \
  requests.post(
    'https://api.planet.com/data/v1/quick-search',
    auth=HTTPBasicAuth(API_KEY, ''),
    json=search_endpoint_request)

results = json.loads(result.text)
for i in results:
    print(i)
    
print(json.dumps(results['features'][3],indent=2))
```
